```python
import pandas as pd

df = pd.read_csv("vdj.csv")
```

```python
# complex.id: 数据记录的复合体ID，用于唯一标识数据记录。
# gene: TCR的基因类型，如TRA（α链）或TRB（β链）。
# cdr3: TCR中CDR3区的氨基酸序列，这是决定TCR特异性的关键区域。
# v.segm: 变异（V）区基因段的名称，对应于TCRα或β链的V区。
# j.segm: 连接（J）区基因段的名称，对应于TCRα或β链的J区。
# species: TCR来源的物种，如HomoSapiens（人类）。
# mhc.a: 主要组织相容性复合体（MHC）分子的A链类型。
# mhc.b: MHC分子的B链类型或其他相关标记（如B2M，β2-微球蛋白）。
# mhc.class: MHC分子的类别，如MHCI或MHCII。
# antigen.epitope: TCR识别的抗原表位的氨基酸序列。
# antigen.gene: 抗原的基因名称。
# antigen.species: 抗原来源的物种，如HIV-1。
# reference.id: 发表这一数据的文献的参考ID，如PMID:15596521。
# method: 数据收集的方法或技术，例如通过tetramer-sort、sanger测序等。
# meta: 元数据，包含有关样本的额外信息，如细胞亚群、捐献者MHC信息、样本来源等。
# cdr3fix: CDR3区序列修正的信息，包括是否需要修正、修正前后的序列等。
# vdjdb.score: 数据记录的质量评分。
# web.method: 数据提交或验证时使用的网页方法。
# web.method.seq: 与web.method相关的序列信息。
# web.cdr3fix.nc: 网络上CDR3修正的非规范（non-canonical）信息。
# web.cdr3fix.unmp: 网络上CDR3修正的未映射（unmapped）信息。
# complex.id: Complex ID of the data record, used to uniquely identify the data record.
# gene: The gene type of the TCR, such as TRA (alpha chain) or TRB (beta chain).
# cdr3: Amino acid sequence of the CDR3 region in the TCR, which is the key region that determines the specificity of the TCR.
# v.segm: Name of the gene segment of the variant (V) region, corresponding to the V region of the TCR alpha or beta chain.
# j.segm: Name of the gene segment of the junction (J) region, corresponding to the J region of the TCR alpha or beta chain.
# species: The species from which the TCR originated, e.g. HomoSapiens (human).
# mhc.a: A-chain type of the major histocompatibility complex (MHC) molecule.
# mhc.b: B-chain type of MHC molecule or other relevant marker (e.g. B2M, β2-microglobulin).
# mhc.class: Class of the MHC molecule, e.g. MHCI or MHCII.
# antigen.epitope: Amino acid sequence of the antigenic epitope recognised by the TCR.
# antigen.gene: The gene name of the antigen.
# antigen.species: The species from which the antigen originated, e.g. HIV-1.
# reference.id: The reference ID of the literature in which this data was published, e.g. PMID:15596521.
# method: The method or technique used for data collection, e.g. by tetramer-sort, sanger sequencing, etc.
# meta: Metadata containing additional information about the sample, e.g. cell subpopulations, donor MHC information, sample source, etc.
# cdr3fix: information about sequence correction of CDR3 region, including whether correction is needed, sequence before and after correction, etc.
# vdjdb.score: Quality score of the data record.
# web.method: The web method used for data submission or validation.
# web.method.seq: Sequence information associated with the web.method.
# web.cdr3fix.nc: Non-canonical information about CDR3 fixes on the web.
# web.cdr3fix.unmp: Unmapped information for CDR3 fixes on the web.
```

```python
# not sure whether we should delete the null-value
df = df.dropna(axis = 0)
df.isnull().sum()
```

Limatation:
    1. Disregarding the length distribution of CDR3.The variations in its length may be associated with specificity.
    2. Unable to capture the correlation among TCR sequences.
    3. When the matrix generated becomes excessively sparse, the computational resources required become too extensive.

Possible solution:
    Embedding layer. The embedding layer can map high-dimensional vectors represented in one-hot encoding to a lower-dimensional continuous space, thereby preserving the semantic information of the sequences while reducing dimensionality.
        嵌入层通过线性变化将高维向量映射到低维连续空间，模型还能学习保留序列之间的关系

```python
df.loc[:, 'species'] = df['species'].astype('category').cat.codes
df.loc[:, 'gene'] = df['gene'].astype('category').cat.codes
df.loc[:, 'mhc.class'] = df['mhc.class'].astype('category').cat.codes
df.loc[:, 'web.method'] = df['web.method'].astype('category').cat.codes
df.loc[:, 'web.method.seq'] = df['web.method.seq'].astype('category').cat.codes
df.loc[:, 'web.cdr3fix.nc'] = df['web.cdr3fix.nc'].astype('category').cat.codes
df.loc[:, 'web.cdr3fix.unmp'] = df['web.cdr3fix.unmp'].astype('category').cat.codes
df.head()
```

```python

# Basic statistics of the dataset
basic_stats = df.describe(include='all')

# Display basic statistics to get an overview of the dataset
basic_stats
```

```python
# gene and antigen.epitope
import matplotlib.pyplot as plt
import seaborn as sns

# Filter out the rows with the most common epitopes for clarity in visualization
top_epitopes = df['antigen.epitope'].value_counts().nlargest(10).index

# Data for visualization
viz_data = df[df['antigen.epitope'].isin(top_epitopes)]

# Plotting gene vs antigen.epitope
plt.figure(figsize=(12, 8))
sns.countplot(data=viz_data, y='antigen.epitope', hue='gene', order=top_epitopes)
plt.title('Distribution of Gene Types for Top 10 Antigen Epitopes')
plt.xlabel('Count')
plt.ylabel('Antigen Epitope')
plt.legend(title='Gene', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

plt.show()
```

```python
# species and antigen.epitope

# Plotting species vs antigen.epitope for top epitopes
plt.figure(figsize=(12, 8))
sns.countplot(data=viz_data, y='antigen.epitope', hue='species', order=top_epitopes)
plt.title('Distribution of Species for Top 10 Antigen Epitopes')
plt.xlabel('Count')
plt.ylabel('Antigen Epitope')
plt.legend(title='Species', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

plt.show()
```

```python
# 探索MHC类别（mhc.class）与抗原表位（antigen.epitope）之间的关系
# Explore the relationship between MHC class (mhc.class) and antigenic epitope (antigen.epitope)

# Plotting mhc.class vs antigen.epitope for top epitopes
plt.figure(figsize=(12, 8))
sns.countplot(data=viz_data, y='antigen.epitope', hue='mhc.class', order=top_epitopes)
plt.title('Distribution of MHC Class for Top 10 Antigen Epitopes')
plt.xlabel('Count')
plt.ylabel('Antigen Epitope')
plt.legend(title='MHC Class', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

plt.show()
```

```python
# We can further explore the relationship between different kinds 
# of MHCs and TCR gene segments (e.g. v.segm and j.segm) and how they affect VDJDB scores
# First, we can create a new column to represent the V and J combination
df['vj_combination'] = df['v.segm'] + '_' + df['j.segm']

# 为了简化分析，我们可能只关注那些最常见的VJ组合
# To simplify the analysis, we may focus only on those VJ combinations that are most common
top_vj_combinations = df['vj_combination'].value_counts().nlargest(10).index

# 筛选出这些组合的数据
# Sift through the data for these combinations
filtered_vj_df = df[df['vj_combination'].isin(top_vj_combinations)]

# 使用散点图来展示VDJDB评分与VJ组合之间的关系，我们也可以按照MHC类别进行着色
# Use a scatterplot to show the relationship between VDJDB scores 
# and VJ combinations, which we can also colour by MHC category
plt.figure(figsize=(16, 10))
sns.scatterplot(x='vj_combination', y='vdjdb.score', hue='mhc.class', data=filtered_vj_df, style='mhc.class', s=100)
plt.title('VDJDB Score by VJ Combination and MHC Class')
plt.xlabel('VJ Combination')
plt.ylabel('VDJDB Score')
plt.xticks(rotation=45)
plt.legend(title='MHC Class')
plt.tight_layout()
```

```python
# TCR CDR3序列长度分布图
# TCR CDR3 sequence length distribution map
df['cdr3_length'] = df['cdr3'].apply(len)
plt.figure(figsize=(10, 6))
plt.hist(df['cdr3_length'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of CDR3 Lengths')
plt.xlabel('CDR3 Length')
plt.ylabel('Frequency')
plt.show()
```

```python
# V和J区段使用频率图
# Frequency map of use of V and J sectors
plt.figure(figsize=(12, 8))
df['v.segm'].value_counts().nlargest(10).plot(kind='bar', color='orange')
plt.title('Top 10 Most Frequent V Segments')
plt.xlabel('V Segment')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(12, 8))
df['j.segm'].value_counts().nlargest(10).plot(kind='bar', color='green')
plt.title('Top 10 Most Frequent J Segments')
plt.xlabel('J Segment')
plt.ylabel('Frequency')
plt.show()
```

```python
# MHC类别与VDJDB评分的关系
# of MHC categories in relation to VDJDB scores
plt.figure(figsize=(10, 6))
boxplot = df.boxplot(column='vdjdb.score', by='mhc.class')
plt.title('VDJDB Score by MHC Class')
plt.suptitle('')  
plt.xlabel('MHC Class')
plt.ylabel('VDJDB Score')
plt.show()
```

```python
# 分析cdr3序列长度与vdjdb.score之间的关系
# Analyse the relationship between cdr3 sequence length and vdjdb.score
# 计算cdr3序列长度
df['cdr3_length'] = df['cdr3'].apply(len)

# 绘制cdr3长度与vdjdb.score的关系图
plt.figure(figsize=(12, 6))
sns.scatterplot(x='cdr3_length', y='vdjdb.score', data=df, alpha=0.6)
plt.title('CDR3 Length vs VDJDB Score')
plt.xlabel('CDR3 Length')
plt.ylabel('VDJDB Score')
plt.show()
```

```python
# 分析v.segm和j.segm的使用频率，以及这些频率如何与VDJDB评分相关
# 绘制v.segm和j.segm使用频率的直方图
# Analyse the frequency of use of v.segm and j.segm and how these frequencies correlate with VDJDB scores
# Plot histograms of v.segm and j.segm usage frequencies
fig, ax = plt.subplots(1, 2, figsize=(16, 6))
sns.countplot(y='v.segm', data=df, order=pd.value_counts(df['v.segm']).iloc[:10].index, ax=ax[0])
ax[0].set_title('Top 10 V Segment Usage')
ax[0].set_xlabel('Frequency')
ax[0].set_ylabel('V Segment')
sns.countplot(y='j.segm', data=df, order=pd.value_counts(df['j.segm']).iloc[:10].index, ax=ax[1])
ax[1].set_title('Top 10 J Segment Usage')
ax[1].set_xlabel('Frequency')
ax[1].set_ylabel('J Segment')
plt.tight_layout()
plt.show()
```

```python
# MHC类别与评分关系
# MHC categories in relation to scores
plt.figure(figsize=(10, 6))
sns.boxplot(x='mhc.class', y='vdjdb.score', data=df)
plt.title('VDJDB Score Distribution Across MHC Classes')
plt.xlabel('MHC Class')
plt.ylabel('VDJDB Score')
plt.show()
```
